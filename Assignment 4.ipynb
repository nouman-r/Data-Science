```python
gdp = pd.read_excel("gdplev.xls", skiprows=5, usecols = [4,6], convert_float=False,\
                    names =('Quarter','GDP chained 2009'))
l= gdp[gdp.Quarter == "2000q1"].index
gdp = gdp.iloc[l[0]:,:]
gdp.reset_index(inplace=True)
gdp.tail()
```

```python
#Question 2
recession_start = []
recession_end = []
for i in list(range(0,64,1)):
    if gdp.loc[i,'GDP chained 2009']>gdp.loc[i+1,'GDP chained 2009'] and \
    gdp.loc[i+1,'GDP chained 2009']>gdp.loc[i+2,'GDP chained 2009']:
        recession_start.append(gdp.loc[i+1,'Quarter'])
    elif len(recession_start) != 0:
            if (gdp.loc[i,'GDP chained 2009']<gdp.loc[i+1,'GDP chained 2009'] and \
    gdp.loc[i+1,'GDP chained 2009']<gdp.loc[i+2,'GDP chained 2009']) and \
            (gdp.loc[i,'GDP chained 2009']<gdp.loc[i-1,'GDP chained 2009'] and \
    gdp.loc[i-1,'GDP chained 2009']<gdp.loc[i-2,'GDP chained 2009']):
                recession_end.append(gdp.loc[i+2,'Quarter'])

            
        
recession_end = ''.join(recession_end)
recession_start = ''.join(recession_start[0])
q_before_rec = "2008q2"
recession_period = gdp.Quarter[(gdp.Quarter >= recession_start) & (gdp.Quarter < recession_end)].tolist()
#gdp.Quarter.apply(lambda x: x>recession_end[0])
gdp[gdp.Quarter.isin(recession_period)]
recession_start
```
